```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

select_drugs=["Capomulin", "Infubinol", "Ketapril", "Placebo"]

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
df_mouse=pd.read_csv(mouse_drug_data_to_load)
df_trial=pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
df_combined=pd.merge(df_mouse,df_trial, on="Mouse ID")
# Display the data table for preview
df_combined
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
df_grouped_mean_volume=df_combined.groupby(["Drug","Timepoint"])["Tumor Volume (mm3)"].mean().reset_index()
# Preview DataFrame
df_grouped_mean_volume
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
df_grouped_sem_volume=df_combined.groupby(["Drug","Timepoint"])["Tumor Volume (mm3)"].sem().reset_index()
# Preview DataFrame
df_grouped_sem_volume

```

```python
# Minor Data Munging to Re-Format the Data Frames
df_reformatted_volume=df_grouped_mean_volume.set_index(["Timepoint","Drug"]).unstack()

# Preview that Reformatting worked
df_reformatted_volume
```

```python
errx=df_grouped_sem_volume[df_grouped_sem_volume["Drug"].isin(select_drugs)]["Timepoint"]
erry=df_grouped_mean_volume[df_grouped_mean_volume["Drug"].isin(select_drugs)]["Tumor Volume (mm3)"]
yerr=df_grouped_sem_volume[df_grouped_sem_volume["Drug"].isin(select_drugs)]["Tumor Volume (mm3)"]
```

```python
# Generate the Plot (with Error Bars)
#plt.plot(df_reformatted.loc[:,("Tumor Volume (mm3)",['Capomulin', 'Infubinol','Ketapril','Placebo'])], color="red")
plt.plot(df_reformatted_volume.loc[:,("Tumor Volume (mm3)",['Capomulin'])], color="red", marker="o",label="Capomulin",linestyle="dotted",linewidth=1)
plt.plot(df_reformatted_volume.loc[:,("Tumor Volume (mm3)",['Infubinol'])], color="blue",marker="^",label="Infubinol",linestyle="dotted",linewidth=1)
plt.plot(df_reformatted_volume.loc[:,("Tumor Volume (mm3)",['Ketapril'])], color="green",marker="s",label="Ketapril",linestyle="dotted",linewidth=1)
plt.plot(df_reformatted_volume.loc[:,("Tumor Volume (mm3)",['Placebo'])], color="black",marker="d",label="Placebo",linestyle="dotted",linewidth=1)
plt.legend()

plt.errorbar(errx,erry,yerr=yerr,fmt='|',color="limegreen")


plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume(mm3)")
plt.grid()
# Save the Figure
plt.show()

```

```python
df_summary=df_grouped_mean_volume.set_index("Timepoint")
df_summary =df_summary.loc[(0,45),:].sort_values(by=["Drug","Timepoint"]).reset_index()
df_summary['pct_ch'] = 100*df_summary.groupby('Drug')['Tumor Volume (mm3)'].pct_change()
df_summary=df_summary[['Timepoint','Drug','pct_ch']].set_index("Timepoint").loc[45,:]
df_summary
```

```python
# Store all Relevant Percent Changes into a Tuple
#passing_tuples = [tuple(x) for x in df_summary.values if ((x[0] == "Capomulin") | (x[0] == "Infubinol")| (x[0] == "Ketapril")| (x[0] == "Placebo"))&(x[1] >= 0)]
#passing_tuples
#passing_drugs=list(zip(*passing_tuples))[0]
#passing_percent_changes = list(zip(*passing_tuples))[1]
#failing_tuples = [tuple(x) for x in df_summary.values if ((x[0] == "Capomulin") | (x[0] == "Infubinol")| (x[0] == "Ketapril")| (x[0] == "Placebo"))&(x[1] < 0)]
#failing_tuples
#failing_drugs=list(zip(*failing_tuples))[0]
#failing_percent_changes = list(zip(*failing_tuples))[1]

passing_tuples = [tuple(x) for x in df_summary.values if (x[0] in select_drugs) &(x[1] >= 0)]
passing_drugs=list(zip(*passing_tuples))[0]
passing_percent_changes = list(zip(*passing_tuples))[1]
failing_tuples = [tuple(x) for x in df_summary.values if (x[0] in select_drugs) &(x[1] < 0)]
failing_drugs=list(zip(*failing_tuples))[0]
failing_percent_changes = list(zip(*failing_tuples))[1]


# Orient widths. Add labels, tick marks, etc. 
x_pos=np.arange(len(passing_drugs)+len(failing_drugs))
fig1=plt.bar(failing_drugs, failing_percent_changes, color='green',width=-1, align='edge')
fig2=plt.bar(passing_drugs, passing_percent_changes, color='red',width=-1, align='edge')
plt.grid()
plt.title("Tumor Change Over 45 Day Treatment")
plt.ylabel("% Tumor Volume Change")
plt.ylim(-30, 70)
plt.xlim(-1.25,3.25)

# Use functions to label the percentages of changes
ax = plt.gca()
def add_value_labels(ax, spacing=5):
    """Add labels to the end of each bar in a bar chart.

    Arguments:
        ax (matplotlib.axes.Axes): The matplotlib object containing the axes
            of the plot to annotate.
        spacing (int): The distance between the labels and the bars.
    """

    # For each bar: Place a label
    for rect in ax.patches:
        # Get X and Y placement of label from rect.
        y_value = rect.get_height()
        x_value = rect.get_x() + rect.get_width() / 2

        # Number of points between bar and label. Change to your liking.
        space = 1
        # Vertical alignment for positive values
        va = 'bottom'

        # If value of bar is negative: Place label below bar
        if y_value < 0:
            # Invert space to place label below
            space *= -1
            # Vertically align label at top
            va = 'top'

        # Use Y value as label and format number with one decimal place
        label = "{:.0f}%".format(y_value)

        # Create annotation
        ax.annotate(label,xy=(x_value,space),textcoords="offset points",ha='center',va=va,color='white')


add_value_labels(ax)

plt.savefig("image.png")


```
